```python
import numpy as np
a = np.array([[1,2,4],[5,8,9],[2,4,2]])
print("Array created\n",a)
#creating array in numpy wowow
```

```python
b = np.zeros((3,4))
print("Array is\n", b)
```

```python
#constant value
c=np.full((3,3),6)
print("All in 6's\n", c)
```

```python
#random array values
d=np.random.random((2,2))
print(d)
```

```python
e=np.arange(0,30,5)
print(e)
```

```python
f=np.array([[1,2,3,4],[6,5,3,2],[5,8,1,2],[10,4,2,1],[5,6,7,4]])
newf=f.reshape(4,5)
print("Newly arranged one\n",newf)
```

```python
flarr=a.flatten()
print("Ori\n", a)
print("\n", flarr)
```

```python
arr=np.array([[1,2,3,4],[3,2,4,2],[3,2,1,2]])
print("\nNo. of dimesnions is", arr.ndim)
print("Shape of array",arr.shape)
print("\nNo. of array", arr.size)
print("\nArray type", arr.dtype)
newtype=arr.astype('f')
print("\nConverted array", newtype)
print("\nConverted type", newtype.dtype)
```

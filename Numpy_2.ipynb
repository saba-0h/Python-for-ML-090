```python
import numpy as np
arr=np.array([[-1,2,0,4],[4,-0.5,6,0],[2.6,0,7,8],[3,-7,4,2.0]])
print("\n", arr)
```

```python
#[start:end:step value]
a=arr[0:3:2]
print("\n",a)
```

```python
#return every other element in the array
arr=np.array([1,2,3,4,5,6,7])
print("\n", arr[::2])
```

```python
arr=np.array([[-1,2,0,4],[4,-0.5,6,0],[2.6,0,7,8],[3,-7,4,2.0]])
temp=arr[:2,:3]
print("\n", temp)
```

```python
#integer array indexing
tem=arr[[0,1,2,3],[3,2,1,0]]
print("\n", tem)
```

```python
#boolen array conditions
con= arr>2
temp=arr[con]
print("\n", temp)
```

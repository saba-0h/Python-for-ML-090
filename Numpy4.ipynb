```python
import numpy as np
arr1=[10,20,30,40,50]
arr2=[2,4,6,8,10]
a=np.array(arr1)
b=np.array(arr2)
print(a,b)#original
print(a+b)#add
print(a-b)#subract
print(a*b)#multiply
print(a/b)#divide
print(a.dot(b))#dot product
sclr=5
print("Result", a*sclr)#scalar multiply
def my_func(x,y):
  if x>y:
    return x-y
  else:
    return x+y
arr1=[10,4,20]
arr2=[2,3,30]
vec_func=np.vectorize(my_func)
print("result", vec_func(arr1,arr2))
```
